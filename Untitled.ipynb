```python
import numpy as np
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
from brewer2mpl import qualitative
```

```python
video = pd.read_csv('vg.csv')
video = video.dropna(axis=0)
video['Year_of_Release']=video['Year_of_Release'].astype(int)
video
```

```python
PS3=video[video['Platform']=='PS3']
PS4=video[video['Platform']=='PS4']
Wii=video[video['Platform']=='Wii']
WiiU=video[video['Platform']=='WiiU']
X360=video[video['Platform']=='X360']
XOne=video[video['Platform']=='XOne']
gm=pd.concat([Wii,WiiU,PS3,PS4,X360,XOne], join='inner')
```

```python
PS31=PS3.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
PS31['platform']='PS3'
PS41=PS4.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
PS41['platform']='PS4'
Wii1=Wii.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
Wii1['platform']='Wii'
WiiU1=WiiU.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
WiiU1['platform']='WiiU'
X3601=X360.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
X3601['platform']='X360'
XOne1=XOne.groupby(['Year_of_Release'],as_index=False)['Year_of_Release'].agg({'number':'count'})
XOne1['platform']='XOne'
```

```python
gm1=pd.concat([PS31,PS41,Wii1,WiiU1,X3601,XOne1], join='inner')
```

```python
sns.lineplot(x="Year_of_Release", y="number", hue="platform",style='platform',markers=True,dashes=False,data=gm1)
```

```python
PS32=PS3.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
PS32['platform']='PS3'
PS42=PS4.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
PS42['platform']='PS4'
Wii2=Wii.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
Wii2['platform']='Wii'
WiiU2=WiiU.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
WiiU2['platform']='WiiU'
X3602=X360.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
X3602['platform']='X360'
XOne2=XOne.groupby(['Year_of_Release'],as_index=False)['Global_Sales'].agg({'number(in millions of units)':'sum'})
XOne2['platform']='XOne'

gm2=pd.concat([PS32,PS42,Wii2,WiiU2,X3602,XOne2], join='inner')
gm2=gm2.rename(columns={'Year_of_Release':'Global sales'})
```

```python
sns.lineplot(x="Global sales", y="number(in millions of units)", hue="platform",style='platform',markers=True,dashes=False,data=gm2)
```

```python
plt.style.use('classic')
ratingsales = gm.groupby(['Rating','Platform']).Global_Sales.sum()
ratingsales.unstack().plot(kind='bar',stacked=True,  colormap= 'Greens', 
                           grid=False)
plt.title('Stacked Barplot of Sales per Rating type ')
plt.ylabel('Sales')
```

```python
plt.style.use('classic')
genreSales = gm.groupby(['Genre','Platform']).Global_Sales.sum()
genreSales.unstack().plot(kind='bar',stacked=True,  colormap= 'Reds', 
                          grid=False)
plt.title('Stacked Barplot of Sales per Game Genre')
plt.ylabel('Sales')
```
